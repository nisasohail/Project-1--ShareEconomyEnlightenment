```python
CityCount=df5['City'].value_counts()
CityCountDF=CityCount.to_frame()
CityCountDF=CityCountDF.rename(columns={"City": "Total_Listings_in_City"})
CityCountDF.to_csv('CityCountDF.csv')
CityCountDF.head()
```

```python
RatePerNight=df5['average_rate_per_night'].tolist()
prices=df5['average_rate_per_night'].str.extract('(\d+)').astype(int)
RatePerNightFloat=prices[0].tolist()
df5['Avg_Rate_Float']= RatePerNightFloat
```

```python
df8=df5.drop(columns=['average_rate_per_night'])
Avg_Rate_by_Zip=df8.groupby(['Zipcode']).mean()
Avg_Rate_by_Zip=Avg_Rate_by_Zip.drop(columns=['Unnamed: 0','id', 'lat', 'lon', 'latitude', 'longitude'])
Avg_Rate_by_Zip.head()
```

```python
Avg_Rate_by_City=df8.groupby(['City']).mean()
Avg_Rate_by_City=Avg_Rate_by_City.drop(columns=['Unnamed: 0','id', 'lat', 'lon','Zipcode', 'latitude', 'longitude'])
Avg_Rate_by_City.to_csv('Avg_Rate_by_City.csv')
Avg_Rate_by_City.head()
```
